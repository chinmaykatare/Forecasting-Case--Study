# Model Comparison & Visualization
Run AFTER train_all.py completes. Compares all model metrics.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys
sys.path.insert(0, '..')

dfs = {}
for name, path in [('XGBoost','../outputs/xgb_metrics.csv'),
                    ('Prophet','../outputs/prophet_metrics.csv'),
                    ('SARIMA', '../outputs/sarima_metrics.csv'),
                    ('LSTM',   '../outputs/lstm_metrics.csv')]:
    if os.path.exists(path):
        tmp = pd.read_csv(path)
        tmp['model'] = name
        dfs[name] = tmp

all_metrics = pd.concat(dfs.values(), ignore_index=True)
all_metrics.head()
```

```python
# Average RMSE per model
summary = all_metrics.groupby('model')[['RMSE','MAE','MAPE%']].mean().round(2)
print(summary)

summary['RMSE'].plot(kind='bar', figsize=(8,4), title='Average RMSE by Model', color='steelblue')
plt.ylabel('RMSE')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../outputs/model_comparison.png', dpi=120)
plt.show()
```

```python
# Best model distribution
best = pd.read_csv('../outputs/best_models.csv')
best['best_model'].value_counts().plot(kind='pie', autopct='%1.0f%%', figsize=(6,6),
                                        title='Best Model per State')
plt.tight_layout()
plt.savefig('../outputs/best_model_pie.png', dpi=120)
plt.show()
print(best)
```
